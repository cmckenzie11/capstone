```python
red, green, blue = cat.copy(), cat.copy(), cat.copy()

red[:, :, (1,2)] = 0
green[:, :, (0,2)] = 0
blue[:, :, (0,1)] = 0
```

```python
#Useful code
#Save for showing images
import matplotlib.pyplot as plt
import numpy as np

def show_images(images,titles=None):
    """Display a list of images"""
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n) # Make subplot
        if image.ndim == 2: # Is image grayscale?
            plt.gray() # Only place in this blog you can't replace 'gray' with 'grey'
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
```

```python
show_images([red,green,blue],['red','green','blue'])
```

```python
#grayscaling
from skimage.color import rgb2gray
gray = rgb2gray(cat)
```

```python
io.imshow(gray)
```

```python
gray.shape
```

```python
from skimage.exposure import equalize_hist
```

```python
equalized = equalize_hist(gray)
io.imshow(equalized)
```

```python
#Lets try blurring the image.
from skimage.filters import gaussian_filter
```

```python
blurred = gaussian_filter(equalized,sigma=2)
blurred2 = gaussian_filter(equalized, sigma=6)
```

```python
show_images([equalized,blurred,blurred2])
```

```python
from skimage.feature import corner_harris, corner_peaks
```

```python
#Handy function for corner detection!
def show_corners(corners,image,title=None):
    """Display a list of corners overlapping an image"""
    fig = plt.figure()
    plt.imshow(image)
    # Convert coordinates to x and y lists
    y_corner,x_corner = zip(*corners)
    plt.plot(x_corner,y_corner,'o',color='r') # Plot corners
    if title:
        plt.title(title)
    plt.xlim(0,image.shape[1])
    plt.ylim(image.shape[0],0) # Images use weird axes
    fig.set_size_inches(np.array(fig.get_size_inches()) * 1.5)
    plt.show()
    print("Number of corners:" + str(len(corners)))
```

```python
cat_corners = corner_peaks(corner_harris(equalized),min_distance=1)
```

```python
show_corners(cat_corners, equalized)
```

```python
#Lets try again with the blurred image, should detect fewer corners?
blurred_corners = corner_peaks(corner_harris(blurred),min_distance=1)
show_corners(blurred_corners, blurred)
```

```python
#Yup, basically doesn't detect any corners on this image.
```

```python
from skimage.filters import sobel_h, sobel_v
```

```python
horizontal = sobel_h(equalized)
vertical = sobel_v(equalized)
show_images([equalized, horizontal, vertical],['normal','horizontal','vertical'])
```
